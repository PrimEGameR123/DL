Normalizing the data ensures that all features contribute equally to the learning process. Without normalization, features with larger scales or variances can dominate the optimization process, which can lead to suboptimal model performance. <br>
The MNIST dataset is made up of 28x28 pixel grayscale images of handwritten digits (0-9). Each pixel in the image has a value between 0 (black) and 255 (white).

```python
# normalizing
x_train = x_train/255
x_test = x_test/255
```

### before creating Neural network, lets understand the input.
* Our input is a $28 \times 28$ matrix of each image which is in 2D format.
* To feed it into input layer, it must be in 1D. e.g., `[Row 1, Row 2, Row 3, .... , Row 28]`
* The shape becomes from $28 \times 28$ => $784$

```python
x_train[0].shape
```

```python
x_train[0].flatten().shape
```

So, we will use flatten layer to do this.

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

> Whenever we have more than one output in classification, we use **softmax** as activation function. Here we have 10 outputs, from 0 to 9. Whichever output has highest probability will be our output.

```python
model.summary()
```

### **Loss Functions used in Categorical NN**
<font color='magenta'> **1. Binary Cross-Entropy (BCE)**<p></font>
Used for binary classification problems where each output is classified into one of two classes (0 or 1).<p> Formula is,
> $\large\text{BCE} = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$
<p> where,

* $y_i$  is the true label (either 0 or 1),
* $\hat{y}_i$  is the predicted probability for class 1. <p> <p>

<font color='magenta'> **2. Categorical Cross-Entropy (CCE)** <p></font>
Used for multi-class classification where the true labels are provided as one-hot encoded vectors. <br>e.g., `[0, 1, 0], [1, 0, 0], [0, 0, 1]` <p>

The formula for Categorical Cross-Entropy is:
> $\large\text{CCE} = -\frac{1}{N} \sum_{i=1}^{N} \sum_{c=1}^{C} y_{i,c} \log(\hat{y}_{i,c})$
<p> where,

* $y_{i,c}$  is the one-hot encoded true label for classes $c$,
* $\hat{y}_{i,c}$ is the predicted probability for classes $c$,
* $C$ is the number of classes. <p><p>

<font color='magenta'> **3. Sparse Categorical Cross-Entropy (SCCE)** <p></font>
Similar to Categorical Cross-Entropy but used when the labels are provided as integer-encoded (not one-hot encoded). <br> e.g, `[2], [3], [4]`. <br>
This approach is more memory-efficient, especially when dealing with a large number of classes, as it avoids the need to convert labels into one-hot vectors.

The formula for Sparse Categorical Cross-Entropy is:
> $\large\text{SCCE} = -\frac{1}{N} \sum_{i=1}^{N} \log(\hat{y}_{i, y_i})$
<p> where,

* $y_i$ is the true label (an integer representing the class),
* $\hat{y}_{i, y_i}$ is the predicted probability for the true class.

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

We can improve this accuracy too, although it is good already, by changing the architecture of ANN.

```python
model2 = Sequential()
model2.add(Flatten(input_shape=(28,28)))
model2.add(Dense(128, activation='relu'))
model2.add(Dense(32, activation='relu')) # adding extra layer
model2.add(Dense(10, activation='softmax'))
```

```python
model2.summary()
```

```python
model2.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history = model2.fit(x_train, y_train, epochs=25, validation_split=0.2) # increasing epochs
```

```python
y_prob2 = model2.predict(x_test)
y_pred2 = y_prob2.argmax(axis=1)
accuracy_score(y_test, y_pred2)
```

Accuracy is reducing due to overfitting here.

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
